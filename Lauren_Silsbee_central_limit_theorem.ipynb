```
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from numpy.random import randint
from numpy import mean
from scipy import stats
from numpy.random import seed
```

```
#Normal Q1
#Create a line chart plotting the sample mean as more and more trials are conducted, up to 1,000 trials.

results = []
for num_trials in range (1,1000):
  trials = np.random.normal (1, 5, num_trials)
  mean_of_trials = trials.mean()
  results.append(mean_of_trials)

normal_df = pd.DataFrame ({'trials' : results})

normal_df.plot()
```

```
#Normal Q2
#Randomly sample the mean 1,000 times, and plot the resulting frequency of sample means as a histogram.

normal_means = [mean(np.random.normal(1, 5, 1000)) for _ in range (1000)]

plt.hist(normal_means)
plt.show()
```

```
#Binomial Q1
#Create a line chart plotting the sample mean as more and more trials are conducted, up to 1,000 trials.

bionomial_results = []
for num_trials in range(1,1000):
  trials = np.random.binomial(1, .5, num_trials)
  mean_of_trials = trials.mean()
  bionomial_results.append(mean_of_trials)

binomial_df = pd.DataFrame({ 'trials' : bionomial_results})

binomial_df.plot()
```

```
#Binomial Q2
#Randomly sample the mean 1,000 times, and plot the resulting frequency of sample means as a histogram.

binomial_means = [mean(np.random.binomial(1, .5, 1000)) for _ in range(1000)]

plt.hist(binomial_means)
plt.show()
```
